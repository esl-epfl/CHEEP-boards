```python
from si_prefix import si_format
current_min = 100e-9 # 100nA
current_max = 100e-3  # 100mA
gain = 1000 
# gain = 200 # INA210 = 200V/V
# gain = 500 # INA211 = 500V/V
# gain = 1000 # INA212 = 1000V/V
# gain = 50 # INA213 = 50V/V
# gain = 100 # INA214 = 100V/V
# gain = 75 # INA215 = 75V/V

supply_voltage = 1.2
shunt = 0.02

vdrop_min = shunt * current_min
vdrop_max = shunt * current_max

vadc_read_min = gain * vdrop_min
vadc_read_max = gain * vdrop_max

adc_voltage_scale = 3.3 
adc_resolution = 24
adc_step = adc_voltage_scale/2**adc_resolution

print("Calculations for Rshunt = {}Ohms & a gain of {}".format(si_format(shunt), gain))
print("ADC step: {}V".format(si_format(adc_step)))

adc_check_max = "[OK]" 
adc_check_min = "[OK]"

if vadc_read_max > adc_voltage_scale:
    adc_check_max = "[Failed: {}V > {}V]".format(si_format(vadc_read_max), si_format(adc_voltage_scale))

if vadc_read_min < adc_step:

    adc_check_min = "[Failed: {}V < {}V]".format(si_format(vadc_read_min), si_format(adc_step))

print("Voltage drop @ {}A = {}V -> ADC reads {}V, {}".format(si_format(current_min), si_format(vdrop_min), si_format(vadc_read_min),adc_check_min))
print("Voltage drop @ {}A = {}V -> ADC reads {}V {}".format(si_format(current_max), si_format(vdrop_max), si_format(vadc_read_max),adc_check_max))

max_current = (adc_voltage_scale/gain)/shunt
min_current = (adc_step/gain)/shunt

print("Min current: {}A".format(si_format(min_current)))
print("Max current: {}A".format(si_format(max_current)))

```
