```python
# Calculations for the adjustable supplies
vfb = 0.55 # from Datasheet
vout_max = 1.2 # spec
vout_min = 0.6 # spec
ifb = 10e-9 # from datasheet
rmax = vout_min / (ifb * 100) # (3) in datasheet
print("Constraint: R1 + R2 + radj < {} ohms".format(rmax))
r1 = 46400 # hand adjusted
r2 = 36000 # hand adjusted
radj = 500000 # hand adjusted
rsum = r1 + r2 + radj
print("R1 + R2 + radj = {} {}".format(rsum, ("too big!" if rsum>rmax else "OK")))

Vmax = vfb*(1+r1/r2) # (2) in datasheet
print("Vmax = {}V".format(Vmax)) 
Vmin = vfb*(1+r1/(r2+radj)) # (2) in datasheet
print("Vmin = {}V".format(Vmin))
```
