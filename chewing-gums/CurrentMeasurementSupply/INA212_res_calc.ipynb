```python
import eseries

Rshunt=eseries.find_nearest(eseries.E24, 10)
Gain_INA=1000
offset=36e-6
MIN_Load_current = 1e-9
MAX_Load_current = 10e-6
GAIN_ADC = 1 
GAIN_ADC_MAX = 4
ADC_precision = 24
ADC_VDD = 3.3
MIN_step_ADC = ADC_VDD / 2**(ADC_precision)

MIN_Input_INA = offset + MIN_Load_current*Rshunt
MIN_Input_INA_wo_offset = MIN_Load_current*Rshunt
MAX_Input_INA = offset + MAX_Load_current*Rshunt

MIN_Input_ADC = MIN_Input_INA * Gain_INA * GAIN_ADC_MAX
MIN_Input_ADC_wo_offset = MIN_Input_INA_wo_offset * Gain_INA * GAIN_ADC_MAX
MAX_Input_ADC = MAX_Input_INA * Gain_INA * GAIN_ADC_MAX

Min_ADC_steps = MIN_Input_ADC_wo_offset / MIN_step_ADC

print("Resitor value={} Ohm".format(Rshunt))
print("Min ADC value={} V, Max ADC value={} V".format(MIN_Input_ADC,MAX_Input_ADC))
print("Number of ADC steps(without offset) for 1nA ={} steps".format(Min_ADC_steps))
```
