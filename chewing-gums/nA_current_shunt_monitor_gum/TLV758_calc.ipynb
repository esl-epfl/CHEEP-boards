```python
import eseries
VFB=0.55
IFB=10e-9
V_target = 3.3
R1=eseries.find_nearest(eseries.E24, 100000)
R2=eseries.find_nearest(eseries.E24, 20000)
VOUT=VFB*(1+R1/R2)
R_Ratio = V_target/VFB-1
Max_value = V_target/(IFB*100)
Add_R = R1+R2
print("R1={}, R2={}, VOUT={}".format(R1,R2,VOUT))
print("Ideal resistance ratio={}".format(R_Ratio))
print("MAX value for R1+R2 ={}".format(Max_value))
print("Current R1+R2={}".format(Add_R))
```
